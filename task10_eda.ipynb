```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


```

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.shape
df.info()
df.describe()
```

```python
missing_percent = df.isnull().mean() * 100
missing_percent
```

```python
df['math score'].hist()
plt.show()

df.boxplot(column='math score')
plt.show()
```

```python
Q1 = df['math score'].quantile(0.25)
Q3 = df['math score'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['math score'] < lower) | (df['math score'] > upper)]
outliers
```

```python
df['outlier_flag'] = ((df['math score'] < lower) | (df['math score'] > upper))
df.head()
```

```python
df['math score'] = np.where(df['math score'] > upper, upper, df['math score'])
df['math score'] = np.where(df['math score'] < lower, lower, df['math score'])
```

```python
df.corr(numeric_only=True)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
